Fuzzy Logic Cross-coupling Controlof Wheeled Mobile Robots

```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
import random
```

```python
"""
PB=positive high
PM=positive medium
PS=positive low
ZO=zero
NB=negetive high
NM=negetive medium
NS=negetive low

Error-In cm
"""
```

```python

error=ctrl.Antecedent(np.arange(-3,3.5,0.5),"error")
error['NS']=fuzzy.trapmf(error.universe,[-3,-3,-2.5,-2])
error['NM']=fuzzy.trimf(error.universe,[-3,-2,-1])
error['NB']=fuzzy.trimf(error.universe,[-2,-1,0])

error['ZO']=fuzzy.trimf(error.universe,[-1,0,1])

error['PS']=fuzzy.trimf(error.universe,[0,1,2])
error['PM']=fuzzy.trimf(error.universe,[1,2,3])
error['PB']=fuzzy.trapmf(error.universe,[2,2.5,3,3])

error.view()
plt.show()
```

```python
change_In_Error=ctrl.Antecedent(np.arange(-3,3.5,0.5),"change_In_Error")
change_In_Error['NS']=fuzzy.trapmf(change_In_Error.universe,[-3,-3,-2.5,-2])
change_In_Error['NM']=fuzzy.trimf(change_In_Error.universe,[-3,-2,-1])
change_In_Error['NB']=fuzzy.trimf(change_In_Error.universe,[-2,-1,0])

change_In_Error['ZO']=fuzzy.trimf(change_In_Error.universe,[-1,0,1])

change_In_Error['PS']=fuzzy.trimf(change_In_Error.universe,[0,1,2])
change_In_Error['PM']=fuzzy.trimf(change_In_Error.universe,[1,2,3])
change_In_Error['PB']=fuzzy.trapmf(change_In_Error.universe,[2,2.5,3,3])

change_In_Error.view()
plt.show()
```

```python
#Contol signal
control_Signal=ctrl.Consequent(np.arange(-3,3.5,0.5),"control_Signal")
control_Signal['NS']=fuzzy.trapmf(control_Signal.universe,[-3,-3,-2.5,-2])
control_Signal['NM']=fuzzy.trimf(control_Signal.universe,[-3,-2,-1])
control_Signal['NB']=fuzzy.trimf(control_Signal.universe,[-2,-1,0])

control_Signal['ZO']=fuzzy.trimf(control_Signal.universe,[-1,0,1])

control_Signal['PS']=fuzzy.trimf(control_Signal.universe,[0,1,2])
control_Signal['PM']=fuzzy.trimf(control_Signal.universe,[1,2,3])
control_Signal['PB']=fuzzy.trapmf(control_Signal.universe,[2,2.5,3,3])

control_Signal.view()
plt.show()
```

```python
# rules
rule1=ctrl.Rule(error['NB'] & change_In_Error['NB'],control_Signal['NB'])
rule2=ctrl.Rule(error['NB'] & change_In_Error['NM'],control_Signal['NB'])
rule3=ctrl.Rule(error['NB'] & change_In_Error['NS'],control_Signal['NB'])
rule4=ctrl.Rule(error['NB'] & change_In_Error['ZO'],control_Signal['NB'])
rule5=ctrl.Rule(error['NB'] & change_In_Error['PS'],control_Signal['NM'])
rule6=ctrl.Rule(error['NB'] & change_In_Error['PM'],control_Signal['NM'])
rule7=ctrl.Rule(error['NB'] & change_In_Error['PB'],control_Signal['NS'])


rule8=ctrl.Rule(error['NM'] & change_In_Error['NB'],control_Signal['NB'])
rule9=ctrl.Rule(error['NM'] & change_In_Error['NM'],control_Signal['NB'])
rule10=ctrl.Rule(error['NM'] & change_In_Error['NS'],control_Signal['NM'])
rule11=ctrl.Rule(error['NM'] & change_In_Error['ZO'],control_Signal['NM'])
rule12=ctrl.Rule(error['NM'] & change_In_Error['PS'],control_Signal['NS'])
rule13=ctrl.Rule(error['NM'] & change_In_Error['PM'],control_Signal['NS'])
rule14=ctrl.Rule(error['NM'] & change_In_Error['PB'],control_Signal['NS'])


rule15=ctrl.Rule(error['NS'] & change_In_Error['NB'],control_Signal['NM'])
rule16=ctrl.Rule(error['NS'] & change_In_Error['NM'],control_Signal['NM'])
rule17=ctrl.Rule(error['NS'] & change_In_Error['NS'],control_Signal['NM'])
rule18=ctrl.Rule(error['NS'] & change_In_Error['ZO'],control_Signal['NS'])
rule19=ctrl.Rule(error['NS'] & change_In_Error['PS'],control_Signal['NS'])
rule20=ctrl.Rule(error['NS'] & change_In_Error['PM'],control_Signal['NS'])
rule21=ctrl.Rule(error['NS'] & change_In_Error['PB'],control_Signal['ZO'])


rule22=ctrl.Rule(error['ZO'] & change_In_Error['NB'],control_Signal['ZO'])
rule23=ctrl.Rule(error['ZO'] & change_In_Error['NM'],control_Signal['ZO'])
rule24=ctrl.Rule(error['ZO'] & change_In_Error['NS'],control_Signal['ZO'])
rule25=ctrl.Rule(error['ZO'] & change_In_Error['ZO'],control_Signal['ZO'])
rule26=ctrl.Rule(error['ZO'] & change_In_Error['PS'],control_Signal['ZO'])
rule27=ctrl.Rule(error['ZO'] & change_In_Error['PM'],control_Signal['ZO'])
rule28=ctrl.Rule(error['ZO'] & change_In_Error['PB'],control_Signal['ZO'])



rule29=ctrl.Rule(error['PS'] & change_In_Error['NB'],control_Signal['ZO'])
rule30=ctrl.Rule(error['PS'] & change_In_Error['NM'],control_Signal['PS'])
rule31=ctrl.Rule(error['PS'] & change_In_Error['NS'],control_Signal['PS'])
rule32=ctrl.Rule(error['PS'] & change_In_Error['ZO'],control_Signal['PS'])
rule33=ctrl.Rule(error['PS'] & change_In_Error['PS'],control_Signal['PM'])
rule34=ctrl.Rule(error['PS'] & change_In_Error['PM'],control_Signal['PM'])
rule35=ctrl.Rule(error['PS'] & change_In_Error['PB'],control_Signal['PM'])


rule36=ctrl.Rule(error['PM'] & change_In_Error['NB'],control_Signal['PS'])
rule37=ctrl.Rule(error['PM'] & change_In_Error['NM'],control_Signal['PS'])
rule38=ctrl.Rule(error['PM'] & change_In_Error['NS'],control_Signal['PS'])
rule39=ctrl.Rule(error['PM'] & change_In_Error['ZO'],control_Signal['PM'])
rule40=ctrl.Rule(error['PM'] & change_In_Error['PS'],control_Signal['PM'])
rule41=ctrl.Rule(error['PM'] & change_In_Error['PM'],control_Signal['PB'])
rule42=ctrl.Rule(error['PM'] & change_In_Error['PB'],control_Signal['PB'])



rule43=ctrl.Rule(error['PB'] & change_In_Error['NB'],control_Signal['PS'])
rule44=ctrl.Rule(error['PB'] & change_In_Error['NM'],control_Signal['PM'])
rule45=ctrl.Rule(error['PB'] & change_In_Error['NS'],control_Signal['PM'])
rule46=ctrl.Rule(error['PB'] & change_In_Error['ZO'],control_Signal['PB'])
rule47=ctrl.Rule(error['PB'] & change_In_Error['PS'],control_Signal['PB'])
rule48=ctrl.Rule(error['PB'] & change_In_Error['PM'],control_Signal['PB'])
rule49=ctrl.Rule(error['PB'] & change_In_Error['PB'],control_Signal['PB'])

Rules=[rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,
        rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,
        rule21,rule22,rule23,rule24,rule25,rule26,rule27,rule28,rule29,rule30,
        rule31,rule32,rule33,rule34,rule35,rule36,rule37,rule38,rule39,rule40,
        rule41,rule42,rule43,rule44,rule45,rule46,rule47,rule48,rule49]
```

```python
x=ctrl.ControlSystem(Rules)
y=ctrl.ControlSystemSimulation(x)
```

```python
i=0
initial_Error=0
while (i<10):
    input_Error = random.uniform(-3,3)    
    error_Change=initial_Error-input_Error
    initial_Error=input_Error



    y.input['error']=input_Error
    y.input['change_In_Error']=error_Change
    y.compute()

    print("Error = ",input_Error)
    print("\nChange In Error = ",error_Change)
    print("\n")

    print("Control Signal = ",y.output['control_Signal'])

    control_Signal.view(sim=y)
    plt.show()


    i+=1
```
